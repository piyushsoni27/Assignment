```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import seaborn as sns

from ipywidgets import widgets, interactive

%matplotlib inline

sales_df = pd.read_excel("African Mobile Data.xlsx", sheetname="SalesData")
```

```python
choice = widgets.Dropdown(
    options = ["Region", "Country"],
    value="Region",
    description='Choice:'
)

def plot_dropdown(choice):
    df2 = sales_df.copy()
    
    if(choice == "Region"):
        sns.pairplot(x_vars="Sales", y_vars="Profit", hue="Region", data=sales_df, size=5, aspect=2,
             kind='reg',plot_kws=dict(fit_reg=True, scatter_kws=dict(alpha=0.5)))
    
    if(choice == "Country"):
        groups = sales_df.groupby("Region")

        western_group = groups.get_group("Western")
        eastern_group = groups.get_group("Eastern")
        middle_group = groups.get_group("Middle")
        southern_group = groups.get_group("Southern")
        northern_group = groups.get_group("Northern")

        fig, big_axes = plt.subplots( figsize=(15.0, 15.0) , nrows=5, ncols=1, sharey=True) 

        for (row, big_ax), (name, group) in zip(enumerate(big_axes), groups):
            big_ax.set_title("Region %s \n" % name, fontsize=16)
            big_ax.tick_params(labelcolor=(1.,1.,1., 0.0), top='off', bottom='off', left='off', right='off')
            big_ax._frameon = False

        j=1;
        for i in range(0,3):
            ax = fig.add_subplot(5,3,j)
            ax.set_title('Plot title ' + str(j))
            j+=1

            if(i<2):
                df = eastern_group.loc[eastern_group.Country.isin(eastern_group.Country.unique()[i*5:(i+1)*5])]

                groups = df.groupby("Country")

                for i, (name, group) in enumerate(groups):
                    c = cm.hot(i/5.,1)
                    x=group.Sales
                    y=group.Profit
                    ax.plot(group.Sales, group.Profit, marker="o", linestyle="", label=name, alpha=0.4, color=c)
                    ax.plot(np.unique(x), np.poly1d(np.polyfit(x, y, 1))(np.unique(x)), color=c, alpha=0.8)

                ax.legend()
                #eastern_group.loc[eastern_group.Country.isin(eastern_group.Country.unique()[i*5:(i+1)*5])].plot("Sales", "Profit",marker='o', linestyle="", ax=ax)
            else:
                df = eastern_group.loc[eastern_group.Country.isin(eastern_group.Country.unique()[i*5:])]

                groups = df.groupby("Country")

                for i, (name, group) in enumerate(groups):
                    c = cm.hot(i/5.,1)
                    x=group.Sales
                    y=group.Profit
                    ax.plot(group.Sales, group.Profit, marker="o", linestyle="", label=name, alpha=0.4, color=c)
                    ax.plot(np.unique(x), np.poly1d(np.polyfit(x, y, 1))(np.unique(x)), color=c, alpha=0.8)

                ax.legend()
                #eastern_group.loc[eastern_group.Country.isin(eastern_group.Country.unique()[i*5:])].plot("Sales", "Profit",marker='o', linestyle="", ax=ax)

        for i in range(0,3):
            ax = fig.add_subplot(5,3,j)
            ax.set_title('Plot title ' + str(j))
            j+=1
            if(i<2):
                df = middle_group.loc[middle_group.Country.isin(middle_group.Country.unique()[i*3:(i+1)*3])]

                groups = df.groupby("Country")

                for i, (name, group) in enumerate(groups):
                    c = cm.hot(i/5.,1)
                    x=group.Sales
                    y=group.Profit
                    ax.plot(group.Sales, group.Profit, marker="o", linestyle="", label=name, alpha=0.4, color=c)
                    ax.plot(np.unique(x), np.poly1d(np.polyfit(x, y, 1))(np.unique(x)), color=c, alpha=0.8)

                ax.legend()
                #middle_group.loc[eastern_group.Country.isin(middle_group.Country.unique()[i*3:(i+1)*3])].plot("Sales", "Profit",marker='o', linestyle="", ax=ax)
            else:
                df = middle_group.loc[middle_group.Country.isin(middle_group.Country.unique()[i*3:])]

                groups = df.groupby("Country")

                for i, (name, group) in enumerate(groups):
                    c = cm.hot(i/5.,1)
                    x=group.Sales
                    y=group.Profit
                    ax.plot(group.Sales, group.Profit, marker="o", linestyle="", label=name, alpha=0.4, color=c)
                    ax.plot(np.unique(x), np.poly1d(np.polyfit(x, y, 1))(np.unique(x)), color=c, alpha=0.8)

                ax.legend()
               # middle_group.loc[eastern_group.Country.isin(middle_group.Country.unique()[i*3:])].plot("Sales", "Profit",marker='o', linestyle="", ax=ax)

        for i in range(0,3):
            ax = fig.add_subplot(5,3,j)
            ax.set_title('Plot title ' + str(j))
            j+=1
            if(i<2):
                df = northern_group.loc[northern_group.Country.isin(northern_group.Country.unique()[i*2:(i+1)*2])]

                groups = df.groupby("Country")

                for i, (name, group) in enumerate(groups):
                    c = cm.hot(i/5.,1)
                    x=group.Sales
                    y=group.Profit
                    ax.plot(group.Sales, group.Profit, marker="o", linestyle="", label=name, alpha=0.4, color=c)
                    ax.plot(np.unique(x), np.poly1d(np.polyfit(x, y, 1))(np.unique(x)), color=c, alpha=0.8)

                ax.legend()
                #northern_group.loc[northern_group.Country.isin(northern_group.Country.unique()[i*2:(i+1)*2])].plot("Sales", "Profit",marker='o', linestyle="", ax=ax)
            else:
                df = northern_group.loc[northern_group.Country.isin(northern_group.Country.unique()[i*2:])]

                groups = df.groupby("Country")

                for i, (name, group) in enumerate(groups):
                    c = cm.hot(i/5.,1)
                    x=group.Sales
                    y=group.Profit
                    ax.plot(group.Sales, group.Profit, marker="o", linestyle="", label=name, alpha=0.4, color=c)
                    ax.plot(np.unique(x), np.poly1d(np.polyfit(x, y, 1))(np.unique(x)), color=c, alpha=0.8)

                ax.legend()
                #northern_group.loc[northern_group.Country.isin(northern_group.Country.unique()[i*2:])].plot("Sales", "Profit",marker='o', linestyle="", ax=ax)

        for i in range(0,3):
            ax = fig.add_subplot(5,3,j)
            ax.set_title('Plot title ' + str(j))
            j+=1
            if(i<2):
                df = southern_group.loc[southern_group.Country.isin(southern_group.Country.unique()[i*2:(i+1)*2])]
                groups = df.groupby("Country")

                for i, (name, group) in enumerate(groups):
                    c = cm.hot(i/5.,1)
                    x=group.Sales
                    y=group.Profit
                    ax.plot(group.Sales, group.Profit, marker="o", linestyle="", label=name, alpha=0.4, color=c)
                    ax.plot(np.unique(x), np.poly1d(np.polyfit(x, y, 1))(np.unique(x)), color=c, alpha=0.8)

                ax.legend()
                #southern_group.loc[southern_group.Country.isin(southern_group.Country.unique()[i*2:(i+1)*2])].plot("Sales", "Profit",marker='o', linestyle="", ax=ax)
            else:
                df = southern_group.loc[southern_group.Country.isin(southern_group.Country.unique()[i*2:])]
                groups = df.groupby("Country")

                for i, (name, group) in enumerate(groups):
                    c = cm.hot(i/5.,1)
                    x=group.Sales
                    y=group.Profit
                    ax.plot(group.Sales, group.Profit, marker="o", linestyle="", label=name, alpha=0.4, color=c)
                    ax.plot(np.unique(x), np.poly1d(np.polyfit(x, y, 1))(np.unique(x)), color=c, alpha=0.8)

                ax.legend()
                #southern_group.loc[southern_group.Country.isin(southern_group.Country.unique()[i*2:])].plot("Sales", "Profit",marker='o', linestyle="", ax=ax)

        for i in range(0,3):
            ax = fig.add_subplot(5,3,j)
            ax.set_title('Plot title ' + str(j))
            j+=1
            if(i<2):
                df = western_group.loc[western_group.Country.isin(western_group.Country.unique()[i*5:(i+1)*5])]
                groups = df.groupby("Country")

                for i, (name, group) in enumerate(groups):
                    c = cm.hot(i/5.,1)
                    x=group.Sales
                    y=group.Profit
                    ax.plot(group.Sales, group.Profit, marker="o", linestyle="", label=name, alpha=0.4, color=c)
                    ax.plot(np.unique(x), np.poly1d(np.polyfit(x, y, 1))(np.unique(x)), color=c, alpha=0.8)

                ax.legend()
                #western_group.loc[western_group.Country.isin(western_group.Country.unique()[i*5:(i+1)*5])].plot("Sales", "Profit",marker='o', linestyle="", ax=ax)
            else:
                df = western_group.loc[western_group.Country.isin(western_group.Country.unique()[i*5:])]
                groups = df.groupby("Country")

                for i, (name, group) in enumerate(groups):
                    c = cm.hot(i/5.,1)
                    x=group.Sales
                    y=group.Profit
                    ax.plot(group.Sales, group.Profit, marker="o", linestyle="", label=name, alpha=0.4, color=c)
                    ax.plot(np.unique(x), np.poly1d(np.polyfit(x, y, 1))(np.unique(x)), color=c, alpha=0.8)

                ax.legend()
                #western_group.loc[western_group.Country.isin(western_group.Country.unique()[i*5:])].plot("Sales", "Profit",marker='o', linestyle="", ax=ax)

        fig.set_facecolor('w')
        plt.tight_layout()
        plt.show()
```

```python
interactive(plot_dropdown, choice=choice)
```
